# Transformar os tweets em uma estrutura Relacional analisável

```python
# importar o módulo pandas para trabalhar com uma visão de tabelas
import pandas as pd

# Atribuir o conteúdo de do primeiro registro (primeirotweet) a um DataFrame
#tabela = pd.DataFrame(primeirotweet)

tabela = pd.DataFrame(primeirotweet, columns =['created_at', 'id', 'id_str', 'text', 'source', 'truncated', 'in_reply_to_status_id', 'in_reply_to_status_id_str', 'in_reply_to_user_id', 'in_reply_to_user_id_str', 'in_reply_to_screen_name', 'user', 'geo', 'coordinates', 'place', 'contributors', 'is_quote_status', 'extended_tweet', 'quote_count', 'reply_count', 'retweet_count', 'favorite_count', 'entities', 'favorited', 'retweeted', 'filter_level', 'lang', 'timestamp_ms'])

tabela
```

```python
# Verificar quais colunas ele retornar
tabela.columns
```

```python
# Agora vamos usar uma nova variavél para criar um data frame 
# buscando apenas os campos de primeiro nivel (root)
# Vamos usar A função reset_index (), ela é usada para gerar um novo 
# DataFrame ou Série com a redefinição do índice.
# Isso é útil quando o índice precisa ser tratado como uma coluna ou 
# quando o índice não tem sentido e precisa ser redefinido para o padrão antes de outra operação

dataframe = pd.DataFrame(primeirotweet, columns =['created_at', 'id', 'id_str', 'text', 'source', 'truncated', 'in_reply_to_status_id', 'in_reply_to_status_id_str', 'in_reply_to_user_id', 'in_reply_to_user_id_str', 'in_reply_to_screen_name', 'user', 'geo', 'coordinates', 'place', 'contributors', 'is_quote_status', 'extended_tweet', 'quote_count', 'reply_count', 'retweet_count', 'favorite_count', 'entities', 'favorited', 'retweeted', 'filter_level', 'lang', 'timestamp_ms','retweeted_status','quoted_status','quoted_status_permalink']).reset_index(drop=True).iloc[:1]
```

```python
# Vamos remover as colunas que não vamos utilizar no nivel root

dataframe.drop(columns=['quote_count',
                        'reply_count',
                        'retweet_count',
                        'favorite_count',
                        'favorited',
                        'retweeted',
                        'user',
                        'retweeted_status',
                        'quoted_status',
                        'quoted_status_permalink',
                        'entities'
                       ],inplace=True)
```

```python
# Criando colunas para armazenar as colunas do segundo nivel (user)

dataframe['user_id'] = primeirotweet['user']['id']
dataframe['user_id_str'] = primeirotweet['user']['id_str']
dataframe['user_screen_name'] = primeirotweet['user']['screen_name']
dataframe['user_location'] = primeirotweet['user']['location']
dataframe['user_description'] = primeirotweet['user']['description']
dataframe['user_protected'] = primeirotweet['user']['protected']
dataframe['user_verified'] = primeirotweet['user']['verified']
dataframe['user_followers_count'] = primeirotweet['user']['followers_count']
dataframe['user_friends_count'] = primeirotweet['user']['friends_count']
dataframe['user_created_at'] = primeirotweet['user']['created_at']
```

```python
# Vamos Criar colunas com os dados de mensão dos usuários (user_mentions)
# vamos criar uma lista vazia
user_mentions = []

# e vamos popular ela com os dados de entities >> user mentions

for i in range(len(primeirotweet['entities']['user_mentions'])):
    
    # vamos criar um dicionário base para armazenar os dados do for
    # vamos fazer uma cópia usando .copy() para não alterar o conteúdo 
    # de primeirotweet caso alterarmos a variavel dicionariobase
    dicionariobase = primeirotweet['entities']['user_mentions'][i].copy()
    
    # Vamos remover a chave indices usando comando pop e passando o parametro none para evitar erros
    dicionariobase.pop('indices',None)
    
    # Agora vamos atribuir o dicionário a um novo data frame
    df = pd.DataFrame(dicionariobase,index=[0]) # em pandas quando o DF for de uma linha temos que infomar o index
    
    # Agora vamos renomear as chaves
    df = df.rename(columns = {
        'screen_name':'entities_screen_name',
        'name':'entities_name',
        'id':'entities_id',
        'id_str':'entities_id_str'
    })
    # e vamos apendar (adicionar) os dados em df
    user_mentions.append(df)
```

```python
# verificando o conteúdo do indice 1 de user_mentions
user_mentions[0]
```

```python
# agora vamos ver como empilhar esses dados usando o concat
# e ignorando o indice

pd.concat(user_mentions,ignore_index=True)
```

```python
# Agora vamos agrupar esses dados no nosso dataframe original
# para isso vamos criar uma nova variavel e apendar o conteúdo de user_mentions

df_agrup = []

# Criar um laço para carregar cada registro de user_mentions no df_agrup

for i in user_mentions:
    df_agrup.append(
        pd.concat([dataframe.copy(),i],axis=1) # indica o eixo onde vamos acrescentar os dados no caso colunas
    )
```

```python
df_agrup[0]
```

```python
# agora vamos ver como empilhar esses dados usando o concat
# e ignorando o indice

pd.concat(df_agrup,ignore_index=True)
```

# Criar Função para realizar o tratamento dos Tweets e armazenar em um DataFrame Pandas

com o conhecimento obtido nas etapas acima vamos sintetizar tudo em uma função para realizar o tratamento de todos os Tweets

```python
# Criando a Função
def transf_tweet(tweet):
    
    # Vamos usar um manipulador para tratar possiveis erros usando try
    try:

        # Setando o DataFrame
        dataframe = pd.DataFrame(tweet, columns =['created_at', 'id', 'id_str', 'text', 'source', 'truncated', 'in_reply_to_status_id', 'in_reply_to_status_id_str', 'in_reply_to_user_id', 'in_reply_to_user_id_str', 'in_reply_to_screen_name', 'user', 'geo', 'coordinates', 'place', 'contributors', 'is_quote_status', 'extended_tweet', 'quote_count', 'reply_count', 'retweet_count', 'favorite_count', 'entities', 'favorited', 'retweeted', 'filter_level', 'lang', 'timestamp_ms','retweeted_status','quoted_status','quoted_status_permalink']).reset_index(drop=True).iloc[:1]

        # Removendor as colunas que não vamos utilizar no nivel root
        dataframe.drop(columns=['quote_count',
                                'reply_count',
                                'retweet_count',
                                'favorite_count',
                                'favorited',
                                'retweeted',
                                'user',
                                'retweeted_status',
                                'quoted_status',
                                'quoted_status_permalink',
                                'entities'
                               ],inplace=True)

        # Criando colunas para armazenar as colunas do segundo nivel (user)
        dataframe['user_id'] = tweet['user']['id']
        dataframe['user_id_str'] = tweet['user']['id_str']
        dataframe['user_screen_name'] = tweet['user']['screen_name']
        dataframe['user_location'] = tweet['user']['location']
        dataframe['user_description'] = tweet['user']['description']
        dataframe['user_protected'] = tweet['user']['protected']
        dataframe['user_verified'] = tweet['user']['verified']
        dataframe['user_followers_count'] = tweet['user']['followers_count']
        dataframe['user_friends_count'] = tweet['user']['friends_count']
        dataframe['user_created_at'] = tweet['user']['created_at']

        # vamos criar uma lista com as menções dos usuários
        user_mentions = []

        # e vamos popular ela com os dados de entities >> user mentions
        for i in range(len(tweet['entities']['user_mentions'])):

            # vamos criar um dicionário base para armazenar os dados do for
            dicionariobase = tweet['entities']['user_mentions'][i].copy()

            # Vamos remover a chave indices usando comando pop e passando o parametro none para evitar erros
            dicionariobase.pop('indices',None)

            # Agora vamos atribuir o dicionário a um novo data frame
            df = pd.DataFrame(dicionariobase,index=[0])

            # Agora vamos renomear as chaves
            df = df.rename(columns = {
                'screen_name':'entities_screen_name',
                'name':'entities_name',
                'id':'entities_id',
                'id_str':'entities_id_str'
            })

            # e vamos apendar (adicionar) os dados em df
            user_mentions.append(df)

        # Agora vamos agrupar os user_mentions no dataframe original
        # para isso vamos criar uma nova variavel e apendar o conteúdo de user_mentions
        df_agrup = []

        # Criar um laço para carregar cada registro de user_mentions no df_agrup
        # e concatenar as novas colunas no DataFrame inicial
        for i in user_mentions:
            df_agrup.append(
                pd.concat([dataframe.copy(),i],axis=1) 
            )

        # Vamos armazenar o conteúdo do DataFrame na variavel Resultado   
        resultado = pd.concat(df_agrup,ignore_index=True)
        
    # Tratando as exceções 
    except:
        return None
    
    # Toda funçao deve retornar alguma coisa
    return resultado
```

```python
# Testando a função para um unico tweet
transf_tweet(parsed_tweets[1])
```

```python
%%time 
# Esse Comando exibe quando ocorreu a última execução de uma célula de código e quanto tempo levou.
# Transfomar todos os tweets e empilhar eles
final_dataframe = [transf_tweet(tweet) for tweet in parsed_tweets]
```

```python
# eliminar posições vazias conforme o except da função
final_dataframe = [i for i in final_dataframe if i is not None]
```

```python
tratado = pd.concat(final_dataframe,ignore_index=True)

tratado
```
